```python
'''
Level 3 – Data Cleaning
Q11. Missing values per column
'''
import pandas as pd 
# Loading  CSV file
df_sales = pd.read_csv("sales_data.csv")  # make sure the file name matches exactly

# Counting  missing values per column
missing_values = df_sales.isnull().sum()
print(missing_values)

```

```python
#Q12. Fill missing Sales with mean
df_sales["Sales"].fillna(df_sales["Sales"].mean(), inplace=True)
```

```python
#Q13. Drop rows where Profit missing
df_sales.dropna(subset=["Profit"])
```

```python
#Q14. Rename Sales → Total_Sales
df_sales.rename(columns={"Sales": "Total_Sales"})
```

```python
#Q15. Convert Date to datetime
df_sales["Date"] = pd.to_datetime(df_sales["Date"])
```

```python
#Level 4 – Grouping & Aggregation
#Q16. Total sales per city
df_sales.groupby("City")["Sales"].sum()
```

```python
#Q17. Avg profit per Region
df_sales.groupby("Region")["Profit"].mean()
```

```python
#Q18. Sum & mean of Sales per Category
df_sales.groupby("Category")["Sales"].agg(["sum", "mean"])
```

```python
#Q19. Count sales per Salesperson
df_sales.groupby("Salesperson").size()
```
